```python
from sympy import *    
```

```python
t = symbols('t')
```

```python
# parametric function

f_1 = cos(t*5) + 1/3
f_2 = sin(t+1)
f_3 = sin(pow(t,2)) * 4

display(f_1)
display(f_2)
display(f_3)
```

```python
# parametric first derivative

f_1_diff = f_1.diff(t)
f_2_diff = f_2.diff(t)
f_3_diff = f_3.diff(t)

display(f_1_diff)
display(f_2_diff)
display(f_3_diff)
```

```python
import numpy as np
```

```python

import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

fs = [f_1, f_2, f_3]

for i in range (3):

    xs = np.linspace(-1, 1, 100)
    ys = [fs[i].subs({t:x}) for x in xs]

    axs[i].plot(xs, ys)

plt.show()
```

```python
def gradient_descent(
        
    gradient, start, learn_rate, n_iter=50, tolerance=1e-06, 
    
    callback = lambda x : print(x)
):
    vector = start

    for _ in range(n_iter):

        diff = -learn_rate * gradient(vector)
        
        if np.all(np.abs(diff) <= tolerance):
            break

        vector += diff
        callback(vector)
    return vector
```

```python
values_2 = []

fs_diff = [f_1_diff, f_2_diff, f_3_diff]

# direct differentiation; 

for f_diff in fs_diff:

    values = []

    gradient_descent(
        
        gradient=lambda v: f_diff.subs({t:v}), 
        start=1.0, learn_rate=0.1,
        callback= lambda x : values.append(x)
    )

    values_2.append(values)

print( values_2 )
```

```python
def plot_grad_vals(fun, values, axs):

    xs = np.linspace(-1, 1, 100)
    ys = [fun(x) for x in xs]

    xs1 = values
    ys1 = [fun(x) for x in values]

    axs.plot(xs, ys)
    axs.plot(xs1, ys1,  '-o', c='r', ls='--')
```

```python

import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

fs = [f_1, f_2, f_3]

for i in range (3):

    plot_grad_vals(lambda x: fs[i].subs({t:x}), values_2[i], axs[i])

plt.show()
```

```python
# symbol definition:

x = symbols('x')

y = symbols('y')
```

```python
# parametric function

f_1 = cos(x*5) + 1/3 + 5*pow(y, 2) - 3*y

f_2 = sin(x+1) + pow(y, 3) - 1/2

f_3 = sin(pow(x,2)) * 4 + 1/2*pow(y, 3) - sin(y)

display(f_1)
display(f_2)
display(f_3)
```

```python
import numpy as np

def gradient_descent(
        
    gradient, start, learn_rate, n_iter=50, tolerance=1e-06, 
    
    callback = lambda x : print(x)
):
    vector = start

    for _ in range(n_iter):
        diff = -learn_rate * gradient(vector)
        if np.all(np.abs(diff) <= tolerance):
            break
        vector += diff
        callback(np.copy(vector))
    return vector
```

```python
values_2 = []

fs = [f_1, f_2, f_3]

for f in fs:

    values = []

    # partial derivatives:

    def grad (v): 
        
        return np.array([

        f.diff(x).subs({x: v[0]}).evalf(), 
        f.diff(y).subs({y: v[1]}).evalf(),
    ], 
    dtype=np.float64)

    gradient_descent(
        gradient= lambda v : grad(v), 
        start=np.array([1.0, 1.0]), 
        learn_rate=0.01,
        n_iter=100,
        callback = lambda x : values.append(x)
    )

    values_2.append(values)

print( np.array( values_2 ).flatten()[:100] )
```

```python
def fun_sym (f_sym, X, Y) : 

    def f(u,v):

        return f_sym.subs({x: u, y : v}).evalf()

    return np.array(
        
        [f(x, y) for (x,y) in zip(np.ravel(X), np.ravel(Y)) ])   
```

```python
def plot_grad_vals_3D(fun, values, ax):

    # Make data.
    X = np.linspace(-1, 1, 10)
    Y = np.linspace(-1, 1, 10)

    X, Y = np.meshgrid(X, Y)
    zs = fun_sym(fun, X, Y)
    Z = zs.reshape(X.shape)

    # Plot the surface.
    ax.plot_surface(X, Y, Z, alpha=0.75)

    x1 = [v[0] for v in values]
    y1 = [v[1] for v in values]
    z1 = [fun.subs({x: v[0], y : v[1]}) for v in values]

    ax.scatter3D(x1, y1, z1, color='red')
    # ax.set_title('3D line plot')
```

```python

import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3, figsize=(15, 5), subplot_kw={"projection": "3d"})

fs = [f_1, f_2, f_3]

for i in range (3):

    plot_grad_vals_3D(fs[i], values_2[i], axs[i])

plt.show()
```

```python
# schema for grad desc with symbolic differentiation:

def symb_diff_grad_strt(f,
                        start, 
                        learn_rate,
                        n_iter):

    values = []

    def grad (v): 
    
        return np.array([

            f.diff(x).subs({x: v[0]}).evalf(), 
            f.diff(y).subs({y: v[1]}).evalf(),
        ], 
        dtype=np.float64)

    return gradient_descent(
        
        gradient    = lambda v : grad(v), 
        start       = start, 
        learn_rate  = learn_rate,
        n_iter      = n_iter,
        callback    = lambda x : values.append(x)

    ), values
```
